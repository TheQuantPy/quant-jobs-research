```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

```python
# List of sample broker websites
brokers = {
    "FXOpen": "https://fxopen.com/en-cy/",
    "FXCM": "https://www.fxcm.com/eu/",
    "Avatrade": "https://www.avatrade.it/",
    "Capital.com": "https://capital.com/en-eu",
    "Fineco": "https://it.finecobank.com/trading/cfd/",
    "RoboMarkets": "https://www.robomarkets.com.cy/eu/",
    "CMC Markets": "https://www.cmcmarkets.com/it-it/",
    "City Index": "https://www.cityindex.com/en-uk/",
    "IG": "https://www.ig.com/it",
    "Eightcap": "https://www.eightcap.eu/en/",
    "Dukascopy": "https://www.dukascopy.com/europe/english/home/",
    "Trade Nation": "https://tradenation.com/en-bs/",
    "HYCM": "",
    "Tickmill": "https://www.tickmill.com/eu",
    "IQ Option": "https://eu.iqoption.com/it",
    "XM": "https://www.xm.com/it",
    "FXPro": "https://www.fxpro.it/",
    "Pepperstone": "https://pepperstone.com/en-eu/",
    "XTB": "https://www.xtb.com/it",
    "OANDA": "https://www.oanda.com/eu-en",
    "CPT Markets": "https://www.cptmarkets.co.uk/",
    "Skilling": "https://skilling.com/eu/it/",
    "Libertex": "https://libertex.com/it",
    "Plus500": "https://www.plus500.com/it",
    "ActivTrades": "https://www.activtrades.eu/en"
}


def fetch_broker_info(url):
    try:
        r = requests.get(url, timeout=10)
        soup = BeautifulSoup(r.content, 'html.parser')
        text = soup.get_text().lower().split('\n')

        loss = next((l.strip() for l in text 
                     if '%' in l and 'lose' in l and 'retail' in l), '')
        leverage = next((l.strip() for l in text 
                         if 'leverage' in l and ':' in l), '')
        margin = next((l.strip() for l in text 
                       if 'margin' in l and '%' in l), '')

        return {"Loss Disclosure": loss, "Leverage Info": leverage, "Margin Info": margin}
    except Exception as e:
        return {"Loss Disclosure": None, "Leverage Info": None, "Margin Info": None}

# Scrape all
data = {name: fetch_broker_info(url) for name, url in brokers.items()}
df = pd.DataFrame.from_dict(data, orient='index').reset_index().rename(columns={'index': 'Broker'})

df
```
