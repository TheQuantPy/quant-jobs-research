```python
from fpdf import FPDF

# Define PDF class
class PDF(FPDF):
    def header(self):
        self.set_font("Arial", "B", 14)
        self.cell(0, 10, "Quantitative Finance Careers - Glossary", ln=True, align="C")
        self.ln(5)

    def chapter_title(self, title):
        self.set_font("Arial", "B", 12)
        self.set_text_color(0)
        self.cell(0, 10, title, ln=True, align="L")
        self.ln(1)

    def chapter_body(self, body):
        self.set_font("Arial", "", 10)
        self.multi_cell(0, 6, body)
        self.ln()

# Create a PDF instance
pdf = PDF()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()

# Role descriptions dictionary
roles = {
    "Quantitative Researcher (also Quantitative Analyst)": """
Also referred to as a buy-side quantitative analyst. Found in hedge funds, proprietary trading firms, and investment banks. 
Typically requires an MSc or PhD (though strong BSc candidates with relevant experience can qualify). 
Researchers develop and test trading strategies using mathematical models and data analysis, aiming to generate alpha.
They often work closely with quant traders and developers. Compensation ranges around $200,000–$250,000+ annually.
Medium work-life balance, stability, and stress; high prestige and demand; career paths can lead to portfolio management.
""",

    "Quantitative Trader (also Algorithmic Trader)": """
Found in hedge funds, trading firms, and banks. Executes and may design trading strategies using quantitative models.
Requires a BSc (Hons) with MSc or PhD desirable. Traders focus on real-time execution and profit-making, 
often collaborating with researchers and developers. Compensation is $200,000–$250,000+.
Environment is high-stress and low-stability, but with high prestige and demand. Career progression tends to be limited.
""",

    "Quantitative Developer (also Quantitative Software Engineer)": """
Implements models and strategies into production systems. Works in trading firms, hedge funds, and banks. 
Requires a BSc or MSc; PhD less common. Compensation ranges from $100,000 (entry) to $200,000+ (experienced).
Works with researchers and traders to turn research into executable code. Moderate prestige, stress, and demand.
Career progression within finance may be limited, but strong outside of finance due to technical skills.
""",

    "Model Validation Quantitative Analyst": """
Also known as a middle or back-office quant. Common in investment and commercial banks. Validates risk and pricing models.
Requires BSc (Hons), MSc or PhD preferred. Entry salaries around $70,000–$80,000, rising to $150,000–$200,000+.
Good work-life balance, high job stability, and medium prestige. Collaborates with compliance but works independently from model developers.
Often focused on identifying weaknesses in existing models.
""",

    "Quantitative Risk Analyst": """
Found in investment and commercial banks, consulting firms. Models and measures financial risks. 
BSc required, MSc or PhD desirable. Salaries range from $70,000–$80,000 (entry) to $150,000–$200,000+.
Low stress, high job stability and demand. Works with risk and compliance teams, occasionally supporting trading desks.
""",

    "Quantitative Portfolio/Investment/Asset Manager": """
An emerging role in asset and portfolio management. Typically filled by experienced buy-side quants.
Applies quantitative models to manage assets. Compensation is $250,000+. Role has low turnover and is highly competitive.
High stress and prestige, but low career progression (top of the ladder). Often seen as the end-goal for many quants.
""",

    "Quantitative Strategist (Strat)": """
Works on trading desks within investment banks. Develops models to support trading, pricing, and risk management.
Requires MSc or PhD. High compensation ($250,000+), stress, and prestige. Works in fast-paced, front-office environments.
Low progression opportunities. Sits between quant researchers and traders, often builds models used in real-time systems.
""",

    "Financial Engineer (also Quantitative Pricing Analyst or Sell-Side Quant)": """
Found in investment banks; focuses on pricing and valuing financial instruments (especially derivatives).
Requires MSc or PhD. High compensation ($250,000+), high stress, and medium work-life balance.
Develops and maintains pricing libraries, often working with sales & trading desks. Sell-side focused.
Low demand due to regulatory saturation, but still competitive.
"""
}

# Add roles to PDF
for title, desc in roles.items():
    pdf.chapter_title(title)
    pdf.chapter_body(desc)

# Save PDF
pdf_path = "Quant_Roles_Glossary_Updated.pdf"
pdf.output(pdf_path)
```
